### Vector Database

-store embeddings in a very fast searchable database

```python
from langchain_community.document_loaders import TextLoader
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma

loaded_document=TextLoader('data/karthikeya.txt').load()
text_splitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=0)
chunks_of_text=text_splitter.split_documents(loaded_document)
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
)

vector_db = Chroma.from_documents(chunks_of_text, embeddings)
print(vector_db)
```

```python
response= vector_db.similarity_search("what he been spending a lot of time learning?")
response[0].page_content
```

### Vector Store as Retriever
-Find the embedding that best answers your question.

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("data/karthikeya.txt")
```

```python
from langchain_community.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loaded_document = loader.load()

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

chunks_of_text = text_splitter.split_documents(loaded_document)
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
) 
vector_db = FAISS.from_documents(chunks_of_text, embeddings)
```

```python
vector_db
```

```python
retriver=vector_db.as_retriever(search_kwargs={"k":2})
```

```python
response=retriver.invoke("what are his projects?")
response
```
